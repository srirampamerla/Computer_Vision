```python
from types import FrameType
#Face Detection
cam =cv2.VideoCapture(0)

face_casade=cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")
while True:
  ret,Frame=cam.read()

  if not ret:
    print("Failed to grab Frame")
    break
  grey=cv2.cvtColor(Frame,cv2.COLOR_BGR2GRAY)
  faces=face_casade.detectMultiScale(grey,scaleFactor=1.1,minNeighbors=7) # sCALE FACTOR -> iT HELPS TO detect the faces at diff scale. we can adjust it based on the size of the face of the present image and for mminimum neighbour -> It helps reduce false +ve's
  #Increasing thevaliuw it will more struict and alod inc the computation also. If it is  ot strict it may miss the face itself

  for (x,y,w,h) in faces:
    cv2.rectangle(Frame,(x,y),(x+w,y+h),(0,255,0),2)

    cv2.putText(Frame,"Face Detected",(x,y-10),cv2.FONT_HERSHEY_SIMPLEX,0.9,(0,255,0),2,cv2.LINE_AA)

  cv2.imshow("Face Detection",Frame)

  if cv2.waitKey(1) & 0xFF == ord("q"):
    break

cam.release()
cv2.destroyAllWindows()
```
