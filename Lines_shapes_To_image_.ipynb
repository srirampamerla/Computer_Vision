If we need to draw any lines or shapes we need canvas

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Create a empty canva
height, width =500,700
blue=(255,127,0) #BGR
canvas=np.full((height,width,3),blue,dtype=np.uint8)

```

```python
plt.imshow(canvas) # BGR it will covert into RGB by default that;'s why we are getting orange color(255 consider as R and 127 coinsdering as G)
```

```python
plt.imshow(cv2.cvtColor(canvas,cv2.COLOR_BGR2RGB)) #plt expects rgb image
plt.show()
```

# Draw a Line


```python
line=cv2.line(canvas,(0,0),(700,100),color=(255,255,255),thickness=5)
```

```python
plt.imshow(line)
```

```python
cv2.line(canvas,(50,50),(650,450),color=(255,255,255),thickness=5)
plt.imshow(cv2.cvtColor(canvas,cv2.COLOR_BGR2RGB))
```

# Draw the Rectangle


```python
# Draw the Rectangle
cv2.rectangle(canvas,(100,100),(200,200),color=(0,255,255),thickness=2) # if thickness we mention -1 it will fill all the shape
plt.imshow(cv2.cvtColor(canvas,cv2.COLOR_BGR2RGB))
```

```python
# Draw the Rectangle
cv2.rectangle(canvas,(100,100),(200,200),color=(0,255,255),thickness=-1) # if thickness we mention -1 it will fill all the shape
plt.imshow(cv2.cvtColor(canvas,cv2.COLOR_BGR2RGB))
```

# Draw Circle

```python
# Draw Circle
#canvas, center point(500,150), raidus=(100) diametr=200 and color()
cv2.circle(canvas,(500,150),100,color=(0,0,255),thickness=2)
plt.imshow(cv2.cvtColor(canvas,cv2.COLOR_BGR2RGB))
```

```python
# Draw Circle
#canvas, center point(500,150), raidus=(100) diametr=200 and color()
cv2.circle(canvas,(500,150),100,color=(0,0,255),thickness=-1)
plt.imshow(cv2.cvtColor(canvas,cv2.COLOR_BGR2RGB))
```

#Draw an Ellipse

```python
#Draw an Ellipse
# canvas, center point/starting point, axes of the ellipse(major,minor),angle of the ellipse, startangle and end angle(where to start from that it will angle),color
cv2.ellipse(canvas,(300,300),(100,50),45,0,360,color=(0,255,0),thickness=2)
plt.imshow(cv2.cvtColor(canvas,cv2.COLOR_BGR2RGB))
```

#  Add Lines,shapes, Text to the image

```python
# create a empty canvas
height=500
width=700
canvas=np.full((height,width,3),(0,0,0),dtype=np.uint8)  #(0,0,0) black
# plt.imshow(cv2.cvtColor(canvas,cv2.COLOR_BGR2RGB))
plt.imshow(canvas)

# Both will give same color in rgb and bgr
```

```python
Text="Hello, Open CV"
font=cv2.FONT_HERSHEY_COMPLEX
#org is the coordinates of the bottom left corner of the text(what needs to start) Means in above text H should start
org=(50,100)
font_scale = 1
color=(255,255,255)
thickness=2

cv2.putText(canvas,Text,org,font,font_scale,color,thickness,cv2.LINE_AA)
plt.imshow(canvas)
```

# Add Multiple fonts

```python
fonts=[

    cv2.FONT_HERSHEY_SIMPLEX,
    cv2.FONT_HERSHEY_COMPLEX,
    cv2.FONT_HERSHEY_SCRIPT_SIMPLEX,
    cv2.FONT_HERSHEY_SCRIPT_COMPLEX,
    cv2.FONT_HERSHEY_COMPLEX_SMALL,
    cv2.FONT_HERSHEY_DUPLEX,
    cv2.FONT_HERSHEY_PLAIN,
    cv2.FONT_HERSHEY_TRIPLEX,
    cv2.FONT_ITALIC

]

y_offset= 50
for i,font in enumerate(fonts):
  text= f"font {i+1}"
  cv2.putText(canvas,text,(50,y_offset),font,1,(255,255,255),5)
  y_offset+=50
plt.imshow(canvas)
```

```python
fonts = [
    ("SIMPLEX", cv2.FONT_HERSHEY_SIMPLEX),
    ("COMPLEX", cv2.FONT_HERSHEY_COMPLEX),
    ("SCRIPT_SIMPLEX", cv2.FONT_HERSHEY_SCRIPT_SIMPLEX),
    ("SCRIPT_COMPLEX", cv2.FONT_HERSHEY_SCRIPT_COMPLEX),
    ("COMPLEX_SMALL", cv2.FONT_HERSHEY_COMPLEX_SMALL),
    ("DUPLEX", cv2.FONT_HERSHEY_DUPLEX),
    ("PLAIN", cv2.FONT_HERSHEY_PLAIN),
    ("TRIPLEX", cv2.FONT_HERSHEY_TRIPLEX),
    ("ITALIC", cv2.FONT_ITALIC),
]

y_offset= 50
for name,font in fonts:
  cv2.putText(canvas,name,(50,y_offset),font,1,(255,255,255),5,cv2.LINE_AA) #cv2.line_AA  Smooth, clear edges
  y_offset+=50
plt.imshow(canvas)
```
