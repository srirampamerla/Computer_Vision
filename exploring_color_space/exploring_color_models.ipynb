```python
import cv2
```

```python
bgr_image=cv2.imread("../mountain.jpg")
```

```python
if bgr_image is None:
    print("Error: Could not read the image.")
else:
    print("Image loaded successfully.")
    print(f"Image shape: {bgr_image.shape}")
    print(f"Image data type: {bgr_image.dtype}")
    print(f"Image size: {bgr_image.size} bytes")
```

```python
cv2.imshow("BGRImage", bgr_image)
cv2.destroyAllWindows()
```

```python
#convert the image into grayscale
gray_image=cv2.cvtColor(bgr_image,cv2.COLOR_BGR2GRAY)

cv2.imwrite("./gray_image.jpg",gray_image)
```

```python
#convert BGR to Hue Saturation Value (HSV) color model

hsv_image=cv2.cvtColor(bgr_image,cv2.COLOR_BGR2HSV)
cv2.imwrite("./hsv_image.jpg",hsv_image)

#HSV is mosty used for calculation doing the threshold of color and other things. So we should not sake for visualizing the image in HSV color model.
```

```python
#convert BGR to Hue Saturation Value (HSV) color model

lab_image=cv2.cvtColor(bgr_image,cv2.COLOR_BGR2LAB)
cv2.imwrite("./lab_image.jpg",lab_image)

#LAB is mosty used for calculation doing the threshold of color and other things. So we should not sake for visualizing the image in HSV color model.
```

```python
#Convert to RGB color model
rgb_image=cv2.cvtColor(bgr_image,cv2.COLOR_BGR2RGB)
cv2.imwrite("./rgb_image.jpg",rgb_image) # we got instead of red color wegot blue color
```

```python
#Convert to RGB color model
rgb_image2=cv2.cvtColor(rgb_image,cv2.COLOR_RGB2BGR)
cv2.imwrite("./rgb_image2.jpg",rgb_image2) #
## bgr format we are unable to see the image in rgb fomrat. But rgB format we convert it is convert to blue
```
