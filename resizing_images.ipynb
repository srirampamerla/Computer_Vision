```python
import cv2
import matplotlib.pyplot as plt
```

```python
image=cv2.imread("/content/giraffe.avif")
```

```python
image
```

```python
image_rgb=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
image_rgb
```

```python
plt.imshow(image_rgb)
```

```python
image_rgb.shape
```

```python
#image_resizing
new_height=400
new_width=400
resized_image=cv2.resize(image_rgb,(new_width,new_height))
plt.imshow(resized_image)
```

```python
image_rgb.shape[1]
```

```python
#image sclaing
scale_percentage=50
s_width=int(image_rgb.shape[1]*scale_percentage/100)
h_width=int(image_rgb.shape[0]*scale_percentage/100)

scaled_image=cv2.resize(image_rgb,(s_width,h_width),interpolation=cv2.INTER_AREA)
print(scaled_image.shape)
plt.imshow(scaled_image)
```

```python
# interpolation

resized_width=600
resized_height=600

#interpolation methods

methods={
    "Nearest":cv2.INTER_NEAREST,
    "Linear":cv2.INTER_LINEAR,
    "Area":cv2.INTER_AREA,
    "Cubic":cv2.INTER_CUBIC,
    "Lanczos":cv2.INTER_LANCZOS4
}
```

```python
resized_images={}
for name,method in methods.items():
  resized_images[name]=cv2.resize(image_rgb,(resized_width,resized_width),interpolation=method)
  cv2.imwrite(f"{name}.jpg",resized_images[name])
```

```python
# Plt the images

plt.figure(figsize=(12,8))
```

```python
#orginal Image
plt.subplot(2,3,1)
plt.title("Original")
plt.imshow(image_rgb)
plt.axis('off')

for i,(name,method) in enumerate(methods.items(),start=2):
  plt.subplot(2,3,i)
  plt.title(name)
  plt.imshow(resized_images[name])
  plt.axis('off')
plt.tight_layout()
plt.show()
```
