```python
import cv2
```

```python
#Read the Image
image=cv2.imread('./mountain.jpg') # This will read the image in matrix grid of pixels and return a numpy array representing the image data.
```

```python
image.shape # This will return the shape of the image in the form of (height, width, channels) where channels represent the color channels (e.g., RGB).

```

```python
#check if image is loaded correctly
if image is None:
    print("Error: Could not read the image.")
else:
    print("Image loaded successfully.")
    print(f"Image shape: {image.shape}")
    print(f"Image data type: {image.dtype}")
    print(f"Image size: {image.size} bytes")
```

```python
#Display the Image
cv2.imshow('Mountain Image', image) # This will open a window to display the image.
cv2.waitKey(0) # This will wait indefinitely until a key is pressed.
cv2.destroyAllWindows() # This will close all OpenCV windows.
```

```python
#Write the image
image_new= image +30 # This will increase the brightness of the image by adding 30 to each pixel value.
cv2.imwrite('./mountain_new.jpg', image_new) # This will save the modified image
```
