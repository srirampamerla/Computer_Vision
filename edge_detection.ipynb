```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image=cv2.imread("/content/giraffe.avif",cv2.IMREAD_GRAYSCALE)
plt.imshow(image,cmap='gray')
```

#Sobel edge detector

```python
#ddepth :Output depth  x derivate in x direction and y derivate in y-direction ksizze =kernel size(it should be odd)
sobel_x=cv2.Sobel(image,ddepth=cv2.CV_64F,dx=1,dy=0,ksize=5)
sobel_y=cv2.Sobel(image,ddepth=cv2.CV_64F,dx=0,dy=1,ksize=5)

sobel_combined=cv2.magnitude(sobel_x,sobel_y)
# sobel_edge=np.uint8(sobel_combined)
plt.imshow(sobel_combined,cmap='gray')
```

#Laplicain

```python
laplicain=cv2.Laplacian(image,ddepth=cv2.CV_64F,ksize=5) # It has both -ve and +ve values
lapician_absolute=cv2.convertScaleAbs(laplicain) # but we need +ve only

```

#canny

```python
#Canny
canny=cv2.Canny(image,threshold1=50,threshold2=100)
```

```python
#Plt the figure

titles=["orginal_image","Sobel","Laplicain Both +ve & -ve ","Laplician_Absolute","Canny"]
images=[image,sobel_combined,laplicain,lapician_absolute,canny]

plt.figure(figsize=(10,10))
for i in range(5):
  plt.subplot(2,3,i+1)
  plt.imshow(images[i],cmap='gray')
  plt.title(titles[i])
  plt.axis('off')
plt.show()
```
