```python
import cv2
```

```python
# Read the video file
video= cv2.VideoCapture('./mountain.mp4')  # Replace 'video.mp4' with your video file path. Video Capture object is used to read the video file.
```

```python
#Display the video
while True: # Loop to read frames from the video. Until the video ends or the user interrupts. While true means it will run indefinitely until a break condition is met.
    ret, frame = video.read()  # Read a frame from the video
    """ret → Boolean (True/False).

True = frame read successfully

False = no frame (end of video or error)

frame → The actual image (frame) captured from the video."""
    if not ret:  # If no frame is returned, break the loop
        break
    # If ret is False, that means the video is finished or something went wrong → exit the loop.

    cv2.imshow('Video Frame', frame)  # Display the frame in a window

    """Opens a window named "Video Frame"

Shows the current frame on screen."""
    if cv2.waitKey(1) & 0xFF == ord('q'):  # Wait for 'q' key to exit  # At time 0 from frame to frame it will wait for 1 millisecond. If the user presses 'q', it will break the loop and exit.
        break

    """Wait 1 millisecond for a key press

Extract only the lowest 8 bits (ASCII part)

Check if the pressed key equals ASCII value of 'q'

If user pressed q, exit the loop"""

# Release the video capture object and close all OpenCV windows
video.release()  # Release the video capture object
cv2.destroyAllWindows()  # Close all OpenCV windows
```

```python
# Write a video 

# Define the codec and create VideoWriter object
fourcc = cv2.VideoWriter_fourcc(*'mp4v')  # Define the codec (mp4v for .mp4 files)

width= int(video.get(cv2.CAP_PROP_FRAME_WIDTH))  # Get the width of the video frames
height= int(video.get(cv2.CAP_PROP_FRAME_HEIGHT))  # Get the height of the video frames
# These properties are used to set the size of the output video.

out = cv2.VideoWriter('output_video.mp4', fourcc, 20.0, (width, height))  # Create VideoWriter object
# 'output_video.mp4' is the output file name, 20.0 is the frames per second, and (640width, 480height) is the frame size.

while True:  # Loop to read frames from the video
    ret, frame = video.read()  # Read a frame from the video
    if not ret:  # If no frame is returned, break the loop
        break

    out.write(frame)  # Write the frame to the output video file
out.release()  # Release the VideoWriter object
video.release()  # Release the video capture object
cv2.destroyAllWindows()  # Close all OpenCV windows
```
