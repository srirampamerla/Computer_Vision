```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image=cv2.imread("/content/filter.jpg",cv2.IMREAD_GRAYSCALE)
```

```python
plt.imshow(image,cmap='gray')
```

```python
horizontal=np.array([[-1,-1,-1],
                     [0,0,0],
                     [1,1,1]])
```

```python
horizontal_filter=cv2.filter2D(image,-1,horizontal)
```

```python
plt.imshow(horizontal_filter,cmap='gray')
```

```python
x=-25
horizontal=np.array([[-1,-1,-1],
                     [x,x,x],
                     [1,1,1]])
horizontal_filter=cv2.filter2D(image,-1,horizontal)
plt.imshow(horizontal_filter,cmap='gray')
```

```python
x=-25
vertical=np.array([[-1,x,1],
                     [-1,x,1],
                     [-1,x,1]])
vertical_filter=cv2.filter2D(image,-1,vertical)
plt.imshow(vertical_filter,cmap='gray')
```

```python
x=-25
diagonal=np.array([[x,-1,-1],
                     [1,x,-1],
                     [1,1,x]])
diagonal_filter=cv2.filter2D(image,-1,diagonal)
plt.imshow(diagonal_filter,cmap='gray')
```

```python
plt.figure(figsize=(20,10))

plt.subplot(2,2,1)
plt.title("Orginal Image")
plt.imshow(image,cmap='gray')
plt.subplot(2,2,2)
plt.title("Horizontal")
plt.imshow(horizontal_filter,cmap='gray')
plt.subplot(2,2,3)
plt.title("Vertical Image")
plt.imshow(vertical_filter,cmap='gray')
plt.subplot(2,2,4)
plt.title("Diagonal Image")
plt.imshow(diagonal_filter,cmap='gray')

plt.axis("off")
plt.tight_layout()
plt.show()
```
