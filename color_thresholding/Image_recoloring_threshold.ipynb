```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image=cv2.imread("/content/giraffe.avif")
```

```python
plt.imshow(image)
```

```python
image_rgb=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
image_hsv=cv2.cvtColor(image,cv2.COLOR_BGR2HSV)
```

```python
plt.imshow(image_rgb)
```

```python
lower_white=np.array([0,0,190])
upper_white=np.array([20,200,255])
```

```python
mask=cv2.inRange(image_hsv,lower_white,upper_white)
```

```python
plt.imshow(mask)
```

```python
black_hsv=np.uint8([[[120,255,0]]])
black_bgr=cv2.cvtColor(black_hsv,cv2.COLOR_HSV2BGR)[0][0]
print(black_bgr)
```

```python
image_res=image.copy()
```

```python
image_res[mask >0]=black_bgr
```

```python
image_result_rgb=cv2.cvtColor(image_res,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(image_result_rgb)
```

```python
fig,axs=plt.subplots(1,3,figsize=(12,6))

axs[0].imshow(image_rgb)
axs[0].set_title("Original")
axs[0].axis("off")
axs[1].imshow(mask)
axs[1].set_title("Mask")
axs[1].axis("off")
axs[2].imshow(image_result_rgb)
axs[2].set_title("Result")
axs[2].axis("off")

plt.tight_layout()
plt.show()
```
